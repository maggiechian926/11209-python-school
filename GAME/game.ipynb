```python
import pygame
import sys
import random

# 初始化Pygame
pygame.init()

# 設定視窗尺寸
window_width, window_height = 300, 300
window = pygame.display.set_mode((window_width, window_height))
pygame.display.set_caption("井字遊戲")

# 設定顏色
white = (255, 255, 255)
black = (0, 0, 0)

# 設定井字遊戲板
board = [[None, None, None], [None, None, None], [None, None, None]]

# 設定井字遊戲狀態
player_turn = 'X'
computer_turn = 'O'
winner = None
draw = False

# 設定字體
font = pygame.font.Font(None, 36)

def check_winner():
    for row in range(3):
        if board[row][0] == board[row][1] == board[row][2] and board[row][0] is not None:
            return board[row][0]
    
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] and board[0][col] is not None:
            return board[0][col]
    
    if board[0][0] == board[1][1] == board[2][2] and board[0][0] is not None:
        return board[0][0]
    
    if board[0][2] == board[1][1] == board[2][0] and board[0][2] is not None:
        return board[0][2]
    
    return None

def check_draw():
    for row in board:
        for cell in row:
            if cell is None:
                return False
    return True

def computer_move():
    # 簡單的電腦AI，隨機選擇一個空單元格下棋
    empty_cells = [(row, col) for row in range(3) for col in range(3) if board[row][col] is None]
    if empty_cells:
        row, col = random.choice(empty_cells)
        board[row][col] = computer_turn

# 遊戲主迴圈
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if winner is None and not draw:
            if player_turn == 'X':
                if event.type == pygame.MOUSEBUTTONDOWN and event.button == 1:
                    x, y = event.pos
                    row, col = y // 100, x // 100
                    if board[row][col] is None:
                        board[row][col] = player_turn
                        winner = check_winner()
                        if winner is None:
                            draw = check_draw()
                        player_turn = 'O'
            else:
                computer_move()
                winner = check_winner()
                if winner is None:
                    draw = check_draw()
                player_turn = 'X'

    window.fill(white)

    for row in range(3):
        for col in range(3):
            left, top = col * 100, row * 100
            rect = pygame.Rect(left, top, 100, 100)
            pygame.draw.rect(window, black, rect, 1)
            cell = board[row][col]
            if cell == 'X':
                x_image = font.render("X", True, black)
                window.blit(x_image, (left + 40, top + 40))
            elif cell == 'O':
                o_image = font.render("O", True, black)
                window.blit(o_image, (left + 40, top + 40))

    if winner:
        text = font.render(f"Player {winner} wins!", True, black)
        window.blit(text, (50, 150))
    elif draw:
        text = font.render("It's a draw!", True, black)
        window.blit(text, (70, 150))

    pygame.display.flip()

# 退出遊戲
pygame.quit()
sys.exit()
```
